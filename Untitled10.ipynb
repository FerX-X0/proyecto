```python
import pandas as pd

df = pd.read_csv("/content/dataset.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
df.duplicated().sum()
```

```python
# Eliminar duplicados
df = df.drop_duplicates()

# Rellenar valores faltantes con promedio por categoría
df["Precio"] = df.groupby("Categoría")["Precio"].transform(lambda x: x.fillna(x.mean()))
df["Descuento"] = df["Descuento"].fillna(df["Descuento"].mean())
df["Ventas"] = df["Ventas"].fillna(df["Ventas"].mean())

# Normalización
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[["Precio","Descuento","Ventas"]] = scaler.fit_transform(df[["Precio","Descuento","Ventas"]])
```

```python
from sklearn.linear_model import LinearRegression

X = df[["Precio", "Descuento", "Mes"]]
y = df["Ventas"]

model = LinearRegression()
model.fit(X, y)

print("Coeficientes:", model.coef_)
print("Intercepto:", model.intercept_)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Barras: ventas por categoría
sns.barplot(x="Categoría", y="Ventas", data=df)
plt.show()

# Dispersión: precio vs ventas
sns.scatterplot(x="Precio", y="Ventas", data=df)
plt.show()

# Boxplot: ventas por mes
sns.boxplot(x="Mes", y="Ventas", data=df)
plt.show()

# Mapa de calor
sns.heatmap(df[['Precio', 'Descuento', 'Mes', 'Ventas']].corr(), annot=True, cmap="coolwarm")
plt.show()
```
